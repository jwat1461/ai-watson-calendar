# Calendar

```python
import calendar
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from datetime import date

def draw_calendar(year, month):
    fig, ax = plt.subplots(figsize=(10, 7))
    ax.set_xlim(0, 7)
    ax.set_ylim(0, 7)
    ax.axis('off')

    # Colors
    header_color = '#4A90D9'
    today_color = '#E74C3C'
    weekend_color = '#F0F4F8'
    cell_color = '#FFFFFF'
    grid_color = '#BDC3C7'

    month_name = calendar.month_name[month]
    ax.text(3.5, 6.7, f'{month_name} {year}',
            ha='center', va='center', fontsize=20, fontweight='bold', color='#2C3E50')

    days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
    for i, day in enumerate(days):
        color = '#E74C3C' if i >= 5 else '#2C3E50'
        rect = mpatches.FancyBboxPatch((i, 5.8), 1, 0.6,
            boxstyle="round,pad=0.02", facecolor=header_color, edgecolor='none')
        ax.add_patch(rect)
        ax.text(i + 0.5, 6.1, day, ha='center', va='center',
                fontsize=11, fontweight='bold', color='white')

    cal = calendar.monthcalendar(year, month)
    today = date.today()

    for week_num, week in enumerate(cal):
        row = 5 - week_num
        for day_num, day in enumerate(week):
            x, y = day_num, row - 0.2
            is_weekend = day_num >= 5
            is_today = (day != 0 and date(year, month, day) == today)

            bg = today_color if is_today else (weekend_color if is_weekend else cell_color)
            rect = mpatches.FancyBboxPatch((x + 0.03, y - 0.75), 0.94, 0.9,
                boxstyle="round,pad=0.03", facecolor=bg, edgecolor=grid_color, linewidth=0.8)
            ax.add_patch(rect)

            if day != 0:
                txt_color = 'white' if is_today else ('#C0392B' if is_weekend else '#2C3E50')
                ax.text(x + 0.5, y - 0.28, str(day),
                        ha='center', va='center', fontsize=13,
                        fontweight='bold' if is_today else 'normal', color=txt_color)

    plt.tight_layout()
    plt.show()

# Show current month
today = date.today()
draw_calendar(today.year, today.month)
```

```python
# View a different month: draw_calendar(year, month)
draw_calendar(2026, 7)  # July 2026
```
